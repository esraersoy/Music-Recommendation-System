```python
#Importing required libraries
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#Requred datasets: user_ratings,song_dataset
data = pd.read_csv("./user_ratings.csv")
database = pd.read_excel("F:/Durham_College_AI/1- semester/Capstone/Song_dataset.xls")
database.drop(["Unnamed: 0"],inplace = True,axis = 1)
data.drop(["Unnamed: 0"],inplace = True,axis = 1)
```

```python
data.fillna(0,inplace = True) 
data
```

```python
#Create user-user similarity matrix using cosine similarity function
user_similarity = cosine_similarity(data)
user_similarity_df = pd.DataFrame(user_similarity,index = data.index,columns = data.index)
user_similarity_df
```

```python
#Gets the most similar user from the matrix
def get_similar_users(user):
    similar_score = user_similarity_df[user]
    similar_score = similar_score.sort_values(ascending = False)
    return similar_score.index[1]
#Returns recommendations from the most similar user 
def get_recommendations(s_user,o_user):
    x = data.iloc[s_user]
    y = data.iloc[o_user]
    s_set = set(x[y!=0].index.values)
    o_set  = set(x[x!=0].index.values)
    return s_set ^ o_set

```

```python
#Main body of program: returns recommedned songs(names and spotify uri)
user = int(input("Enter the target user:"))
similar_user = get_similar_users(user)
recommendations = list(get_recommendations(similar_user,user))
r_song_name = database.iloc[recommendations].Tracks.values
r_artist_name = database.iloc[recommendations].Artist.values
r_uri =  database.iloc[recommendations].uri.values
#print(recommendations)
print(similar_user)
print(recommendations)
print(r_song_name + " by " + r_artist_name + " :" + r_uri )
```
